# stage 2 - Data Cleaning and Pre-Processing

```python
# 1.Missing Values Check
print("Missing Values (Before):")
print(df.isnull().sum())

```

```python
#.2. Handle Missing Values (Imputation)
# Numerical columns fill with mean
num_cols=df.select_dtypes(include=['int64','float64']).columns
df[num_cols]=df[num_cols].fillna(df[num_cols].mean())

```

```python
# categorical columns fill with mode
cat_cols=df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0],inplace=True)
```

```python
# Check for missing values after handling
print("\nMissing Value  (After):")
print(df.isnull().sum())
```

```python
# 3. Handle Duplicates
print("\nDuplicate Rows (Before):", df.duplicated().sum())
```

```python
# Remove duplicates
df = df.drop_duplicates()

print("Duplicate Rows (After):", df.duplicated().sum())
```

```python
df = df.fillna("Not Available")
print(df.head())
```

# lineplot chart


```python
import seaborn as sns

# Convert Registration Date to datetime
df['Date of Incorporation (incorporation_date)'] = pd.to_datetime(df['Date of Incorporation (incorporation_date)'])

# Count registrations by year
yearly = df['Date of Incorporation (incorporation_date)'].dt.year.value_counts().sort_index()

# Plot
plt.figure(figsize=(8,5))
plt.plot(yearly.index, yearly.values, marker='3')
plt.title('Year-wise LLP Registrations')
plt.xlabel('Year')
plt.ylabel('Number of Registrations')
plt.grid(True)
plt.show()
```

# Grouped Analysis

```python
import pandas as pd
# Total Partners by State
result = df.groupby('State Name (state_name)')['Total Partners (total_partners)'].sum()
print(result)

# Average Obligation Contribution by State
result = df.groupby('State Name (state_name)')['Obligation Contribution (obligation_of_contribution)'].mean()
print(result)

# Multiple Aggregations
result = df.groupby('State Name (state_name)').agg({
    'Total Partners (total_partners)': ['sum', 'mean', 'max'],
    'Obligation Contribution (obligation_of_contribution)': ['sum', 'mean']})
print(result)
```

# Grouped Analysis-Grouped analysis is used to summarize data by categories

**Explanation**:Grouped analysis is used to summarize data by categories. Here, the data is grouped by state to calculate the total partners, average obligation contribution, and the number of LLP registrations. This helps compare the performance of different states.

**Showing**:Grouped analysis is a technique that groups data by a category and applies aggregate functions such as sum, mean, count, maximum, or minimum to summarize the data.



# Pivot table

```python
pivot = pd.pivot_table(df,
    values='Obligation Contribution (obligation_of_contribution)',
    index='State Name (state_name)',
    aggfunc='mean')
print(pivot)
```

# Pivot table - Pivot table summarizes the dataset
 **Explanation**:A pivot table summarizes the dataset by grouping data based on state. It calculates values such as the total partners, average obligation contribution, or the number of LLPs. This makes it easy to compare information across different states.
  
**Showing**:analyze large datasets by grouping data and applying aggregate functions such as sum, average, count, maximum, and minimum.
